#### outliers Treatment
Outlier Treatment is important step since K-Mean algorithm uses mean to update the cluster centers and outlier badly effect the mean.

```python
# This Function will Impute the oulier with P99 and P01
def outliers_treat(x):
    x=x.clip_upper(x.quantile(0.99))
    x=x.clip_lower(x.quantile(0.01))
    return x
```

```python
num_var=num_var.apply(outliers_treat)
```

```python
num_var.apply(var_summary).T
```

### Finding Optimal number of clusters

### Silhouette Coefficient 
Silhouette Coefficient will help us to identify the best cluster solution its scores based on¬†how different¬†the clusters are from each other. SC ranges from 1 to -1. Near to 1 is the best solution.
The Silhouette Coefficient is calculated using the mean intra-cluster distance ( a ) and the mean nearest-cluster distance ( b ) for each sample. The Silhouette Coefficient for a sample is (b - a) / max(a, b)

```python
from sklearn import metrics
```

```python
k_range = range(3, 9)
scores = []
for k in k_range:
    km = KMeans(n_clusters=k, random_state=1)
    km.fit(card_scaled2)
    scores.append(metrics.silhouette_score(card_scaled2, km.labels_))
```

```python
scores 
```

Cluster 3 Sloution has given us the max SC value we can select Cluster 3 solution
